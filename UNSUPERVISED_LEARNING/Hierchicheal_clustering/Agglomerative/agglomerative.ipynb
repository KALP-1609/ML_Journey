```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df_iris = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
df_iris.head()
```

## STANDARIZATION

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(df_iris)
```

```python
X_scaled
```

## FEATURE SCALING

To do feature scaling we will apply PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

### PLOTTING

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=load_iris().target)
```

## AGLOMERATIVE CLUSTERING

First we will construct a Dendogram and then we will plot it using the scipy library

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(30,10))
plt.title("Dendogram")
dend = sc.dendrogram(sc.linkage(X_pca,method='ward'))
plt.xlabel('Sample index')
plt.ylabel('Euclidean distances')
plt.axhline(y=15,color='r',linestyle='--')
plt.show()
```

Here we see that when we choose the longest vertical line that is the blue one on the right no horizontal line passes through it , so we will draw a red horizontal line and then we see that the line is intersecting at only 2 points so we will choose K as 2

### IMPLEMENTING AGLOMERATIVE CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
y_pred = ac.fit_predict(X_pca)
y_pred
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=y_pred)
```
